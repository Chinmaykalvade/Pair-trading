```python
# Step 1: Install necessary libraries
!pip install yfinance tensorflow pandas_datareader -q

# Step 2: Import Libraries
import pandas as pd
import numpy as np
import yfinance as yf
from pandas_datareader import data as pdr
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, concatenate
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, mean_absolute_error
from datetime import datetime, timedelta
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

# Suppress warnings for cleaner output
tf.get_logger().setLevel('ERROR')
warnings.filterwarnings('ignore', category=FutureWarning)
pd.options.mode.chained_assignment = None # Suppress pandas slicing warnings

print("Libraries imported successfully.")

# Step 3: Get S&P 500 Tickers
print("\nFetching list of S&P 500 tickers from Wikipedia...")
try:
    sp500_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    sp500_df = pd.read_html(sp500_url)[0]
    # Correctly handle ticker symbols with dots (e.g., BRK.B -> BRK-B)
    tickers = [ticker.replace('.', '-') for ticker in sp500_df['Symbol'].tolist()]
    print(f"Successfully fetched {len(tickers)} tickers.")
except Exception as e:
    print(f"Could not fetch S&P 500 tickers, using a fallback list. Error: {e}")
    tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'NVDA', 'JPM', 'V', 'JNJ', 'UNH'] # Fallback list

# For demonstration, limit the number of tickers to speed up the process.
# To run on the full list, comment out the next line.
TICKER_LIMIT = 3000 # Limiting to 50 for quicker demonstration with more data fetching
print(f"--- NOTE: Limiting run to {TICKER_LIMIT} tickers for demonstration purposes. ---")
tickers = tickers[:TICKER_LIMIT]


# Step 4: Fetch Historical and Fundamental Data (This will take a long time)
end_date = datetime.now()
start_date = end_date - timedelta(days=1500) # Approx. 4 years of data

# Fetch all historical data in one efficient call
print(f"\nFetching historical data for {len(tickers)} tickers...")
hist_data = yf.download(tickers, start=start_date, end=end_date, threads=True)
print("Historical data download complete.")


all_stock_data = {}
failed_tickers = []
print(f"\nFetching fundamental data and assembling final dataset...")

# Collect all possible info keys first to ensure consistent columns
# This helps in creating a DataFrame with all possible fundamental columns
all_info_keys = set()
for ticker in tickers:
    try:
        stock_info_temp = yf.Ticker(ticker).info
        if stock_info_temp:
            all_info_keys.update(stock_info_temp.keys())
    except Exception as e:
        print(f"   -> Could not get info keys for {ticker}: {e}")

# Filter for potentially numerical keys (heuristic)
potential_numeric_info_keys = [
    k for k in all_info_keys if isinstance(yf.Ticker(tickers[0]).info.get(k), (int, float)) or \
    (isinstance(yf.Ticker(tickers[0]).info.get(k), str) and yf.Ticker(tickers[0]).info.get(k).replace('.', '', 1).isdigit()) # Basic check for numeric strings
]
print(f"Discovered {len(potential_numeric_info_keys)} potential numeric fundamental features.")


for i, ticker in enumerate(tickers):
    print(f"Processing ({i+1}/{len(tickers)}): {ticker}")
    try:
        # Extract this ticker's history from the bulk download
        hist = hist_data.loc[:, (slice(None), ticker)]
        hist.columns = hist.columns.droplevel(1) # Remove the ticker level from column names
        hist = hist.dropna() # Drop any rows with NaNs for this specific ticker

        if hist.empty or len(hist) < 300: # Need enough data for sequences
            print(f"   -> Skipping {ticker} (insufficient historical data)")
            failed_tickers.append(ticker)
            continue

        # Fetch fundamental data (this still happens individually)
        stock_info = yf.Ticker(ticker).info

        # A simple check for valid fundamental data
        # Ensure at least 'trailingPE' or 'marketCap' is present as a basic check
        if not stock_info or (stock_info.get('trailingPE') is None and stock_info.get('marketCap') is None):
            print(f"   -> Skipping {ticker} (no fundamental data or key metric missing)")
            failed_tickers.append(ticker)
            continue

        all_stock_data[ticker] = {'history': hist, 'info': stock_info}
    except Exception as e:
        print(f"   -> ERROR processing data for {ticker}: {e}")
        failed_tickers.append(ticker)

print(f"\nFinished data fetching and assembly.")
print(f"Successfully processed: {len(all_stock_data)} tickers.")
print(f"Failed or skipped: {len(failed_tickers)} tickers: {failed_tickers}")

# Step 5: Consolidate and Preprocess Data
if not all_stock_data:
    raise ValueError("No data was successfully fetched. Cannot proceed.")

# --- Identify Static (Fundamental) and Time-Series Features ---
# Consolidate all static features into a single DataFrame
static_features_list = []

# Dynamically collect all numeric fundamental features from the info dictionary
# Ensure consistent columns by using the potential_numeric_info_keys found earlier
for ticker, data in all_stock_data.items():
    info = data['info']
    row_data = {'symbol': ticker}
    for feature in potential_numeric_info_keys:
        value = info.get(feature)
        # Attempt to convert to numeric, set to NaN if conversion fails
        if isinstance(value, (int, float)):
            row_data[feature] = value
        elif isinstance(value, str):
            try:
                row_data[feature] = float(value)
            except ValueError:
                row_data[feature] = np.nan
        else:
            row_data[feature] = np.nan
    static_features_list.append(row_data)

static_df_raw = pd.DataFrame(static_features_list).set_index('symbol')

# --- Clean Static Features ---
# 1. Select only numeric features (redundant after manual conversion, but good for safety)
static_df_numeric = static_df_raw.select_dtypes(include=np.number)

# 2. Drop columns with high percentage of missing values or zero variance
max_nan_perc = 0.4 # Drop columns with more than 40% NaN
min_valid_count = int(len(static_df_numeric) * (1 - max_nan_perc))
static_df_filtered = static_df_numeric.dropna(axis=1, thresh=min_valid_count)

# Drop columns with near-zero variance (constant or almost constant values)
# Calculate variance and remove if below a threshold (e.g., 1e-6)
variances = static_df_filtered.var()
constant_columns = variances[variances < 1e-6].index
static_df_filtered = static_df_filtered.drop(columns=constant_columns)
print(f"Dropped {len(constant_columns)} static features due to near-zero variance.")


# 3. Impute remaining missing values
imputer = SimpleImputer(strategy='median')
static_df_imputed = pd.DataFrame(imputer.fit_transform(static_df_filtered),
                                 index=static_df_filtered.index,
                                 columns=static_df_filtered.columns)
# 4. Scale static features
scaler_static = StandardScaler()
static_df_scaled = pd.DataFrame(scaler_static.fit_transform(static_df_imputed),
                                 index=static_df_imputed.index,
                                 columns=static_df_imputed.columns)

final_static_features = static_df_scaled.columns.tolist()
print(f"\nProcessed static data. Kept {len(final_static_features)} out of {len(static_df_numeric.columns)} numeric features (after filtering).")
print(f"Sample of final static features: {final_static_features[:5]} ... {final_static_features[-5:]}")


# --- Define Time-Series Features ---
# Get all available numerical columns from a sample historical data
sample_ticker = list(all_stock_data.keys())[0]
sample_hist_df = all_stock_data[sample_ticker]['history']

# Dynamically identify all numeric columns in historical data
# Exclude 'Datetime' if it somehow sneaks in and any non-numeric objects
ts_features = sample_hist_df.select_dtypes(include=np.number).columns.tolist()

# Ensure 'Adj Close' is used as the target and remove 'Close' if 'Adj Close' is present
target_col_name = 'Adj Close' if 'Adj Close' in ts_features else 'Close'
if 'Close' in ts_features and target_col_name == 'Adj Close':
    ts_features.remove('Close')

# Ensure the target column is indeed in the features, if not, choose another if possible
if target_col_name not in ts_features and 'Close' in ts_features:
    target_col_name = 'Close'
elif target_col_name not in ts_features:
    raise ValueError(f"Neither 'Adj Close' nor 'Close' found in time-series features for target.")

print(f"\nUsing time-series features: {ts_features}")
print(f"Target variable for prediction: {target_col_name}")


# Create a scaler for time-series data (we'll apply it per-stock)
scaler_ts = MinMaxScaler(feature_range=(0, 1))
# And a scaler for the target variable
scaler_target = MinMaxScaler(feature_range=(0, 1))

# Step 6: Create Sequences for the Multi-Input Model
lookback = 60
X_lstm, X_mlp, y_target = [], [], []

print("\nCreating data sequences for the model...")
skipped_sequence_tickers = []
for ticker, data in all_stock_data.items():
    # Skip if ticker is not in the processed static_df_scaled
    if ticker not in static_df_scaled.index:
        skipped_sequence_tickers.append(ticker)
        continue

    hist = data['history'].copy()
    static_vec = static_df_scaled.loc[ticker].values

    # Ensure all selected ts_features are present in hist
    if not all(feature in hist.columns for feature in ts_features):
        print(f"   -> Skipping {ticker} (missing some time-series features {ts_features} for sequence creation)")
        skipped_sequence_tickers.append(ticker)
        continue

    # Ensure target column is present
    if target_col_name not in hist.columns:
        print(f"   -> Skipping {ticker} (target column '{target_col_name}' not found for sequence creation)")
        skipped_sequence_tickers.append(ticker)
        continue

    # Handle Dividends and Stock Splits carefully: fill NaNs with 0 before scaling if they appear
    # These are usually sparse and might have NaNs for days without dividends/splits
    for col in ['Dividends', 'Stock Splits']:
        if col in hist.columns:
            hist[col] = hist[col].fillna(0)

    # Use a temporary scaler for each stock to avoid data leakage and scale individually
    temp_scaler_ts = MinMaxScaler(feature_range=(0, 1))
    hist_scaled = temp_scaler_ts.fit_transform(hist[ts_features])

    # Scale the target column
    temp_scaler_target = MinMaxScaler(feature_range=(0, 1))
    target_scaled = temp_scaler_target.fit_transform(hist[[target_col_name]])

    for i in range(lookback, len(hist_scaled)):
        # LSTM input: sequence of historical data
        X_lstm.append(hist_scaled[i-lookback:i, :])
        # MLP input: the static feature vector for this stock
        X_mlp.append(static_vec)
        # Target: the next day's scaled closing price
        y_target.append(target_scaled[i, 0])

if skipped_sequence_tickers:
    print(f"Skipped {len(skipped_sequence_tickers)} tickers during sequence creation due to missing data or static feature mismatch.")

X_lstm, X_mlp, y_target = np.array(X_lstm), np.array(X_mlp), np.array(y_target)
print(f"Total samples created: {len(X_lstm)}")
if len(X_lstm) == 0:
    raise ValueError("Sequence creation failed. Check data length and lookback period, or data availability.")

# --- Split into Training and Testing sets ---
# We'll just take the last N samples as a test set for simplicity
test_size = int(len(X_lstm) * 0.2)
train_size = len(X_lstm) - test_size

X_train_lstm, X_test_lstm = X_lstm[:train_size], X_lstm[train_size:]
X_train_mlp, X_test_mlp = X_mlp[:train_size], X_mlp[train_size:]
y_train, y_test = y_target[:train_size], y_target[train_size:]

print(f"Training set size: {train_size}")
print(f"Testing set size: {test_size}")
print(f"LSTM input shape: {X_train_lstm.shape}")
print(f"MLP input shape: {X_train_mlp.shape}")

# Step 7: Build and Train the Multi-Input Model
# --- LSTM Branch for Time-Series Data ---
lstm_input = Input(shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]), name='lstm_input')
lstm_layer = LSTM(units=50, return_sequences=True)(lstm_input)
lstm_dropout = Dropout(0.2)(lstm_layer)
lstm_layer_2 = LSTM(units=50)(lstm_dropout)
lstm_output = Dropout(0.2)(lstm_layer_2)

# --- MLP Branch for Static Data ---
mlp_input = Input(shape=(X_train_mlp.shape[1],), name='mlp_input')
mlp_layer = Dense(units=32, activation='relu')(mlp_input)
mlp_dropout = Dropout(0.2)(mlp_layer)
mlp_output = Dense(units=16, activation='relu')(mlp_dropout)

# --- Concatenate Branches ---
combined_input = concatenate([lstm_output, mlp_output])
final_dense = Dense(units=32, activation='relu')(combined_input)
final_output = Dense(units=1, name='output')(final_dense)

model = Model(inputs=[lstm_input, mlp_input], outputs=final_output)
model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# --- Train the model ---
print("\nTraining the multi-input model...")
history = model.fit(
    [X_train_lstm, X_train_mlp],
    y_train,
    epochs=10, # Keeping epochs low for this demonstration
    batch_size=64,
    validation_split=0.1,
    verbose=1
)
print("Model training complete.")


# Step 8: Evaluate and Analyze
# --- Make Predictions ---
predictions_scaled = model.predict([X_test_lstm, X_test_mlp])
# Re-fit scaler_target_for_inverse on the full range of y_target values for accurate inverse transform
scaler_target_for_inverse = MinMaxScaler(feature_range=(0, 1))
scaler_target_for_inverse.fit(y_target.reshape(-1, 1))

predictions = scaler_target_for_inverse.inverse_transform(predictions_scaled)
y_test_actual = scaler_target_for_inverse.inverse_transform(y_test.reshape(-1, 1))

# --- Calculate Accuracy Metrics ---
rmse = np.sqrt(mean_squared_error(y_test_actual, predictions))
mae = mean_absolute_error(y_test_actual, predictions)
print(f"\nModel Evaluation on Test Set:")
print(f"Root Mean Squared Error (RMSE): ${rmse:.2f}")
print(f"Mean Absolute Error (MAE): ${mae:.2f}")

# --- Feature Importance (Simplified Permutation) ---
print("\nCalculating simplified feature importance...")
baseline_error = mean_squared_error(y_test, predictions_scaled)
importances = {}

# Test all static features
print(f"Testing all {len(final_static_features)} static features for importance.")
for i, name in enumerate(final_static_features):
    # Find the index of the feature in the actual X_train_mlp columns
    try:
        col_idx = static_df_scaled.columns.get_loc(name)
        X_test_mlp_permuted = X_test_mlp.copy()
        np.random.shuffle(X_test_mlp_permuted[:, col_idx])
        permuted_error = mean_squared_error(y_test, model.predict([X_test_lstm, X_test_mlp_permuted], verbose=0))
        importances[name] = permuted_error - baseline_error
    except KeyError:
        print(f"Warning: Feature '{name}' not found in static_df_scaled columns for importance calculation (this shouldn't happen).")

# Test all time-series features
print(f"Testing all {len(ts_features)} time-series features for importance.")
for i, name in enumerate(ts_features):
    X_test_lstm_permuted = X_test_lstm.copy()
    np.random.shuffle(X_test_lstm_permuted[:, :, i]) # Shuffle along the feature axis
    permuted_error = mean_squared_error(y_test, model.predict([X_test_lstm_permuted, X_test_mlp], verbose=0))
    importances[name] = permuted_error - baseline_error

importance_df = pd.DataFrame.from_dict(importances, orient='index', columns=['Importance'])
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Step 9: Visualize Results
print("\nVisualizing top 20 most important features...")
plt.figure(figsize=(12, 10))
sns.barplot(x=importance_df.Importance.head(20), y=importance_df.index[:20], palette='viridis')
plt.title('Top 20 Feature Importances for Price Prediction (All Available Features)')
plt.xlabel('Importance (Increase in MSE after Permutation)')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

print("\nVisualizing predictions for a sample of the test set...")
plt.figure(figsize=(16, 8))
plt.title('Actual vs. Predicted Prices (Sample from Test Set)')
plt.plot(y_test_actual[:500], label='Actual Price', color='blue')
plt.plot(predictions[:500], label='Predicted Price', color='red', linestyle='--')
plt.ylabel('Stock Price (USD)')
plt.xlabel('Time (Sample Index)')
plt.legend()
plt.show()

print("\n--- Analysis Complete ---")
```
